## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Copy to avoid modifying original
df_cleaned = df.copy()

# 1. Convert 'Timestamp' to datetime
df_cleaned['Timestamp'] = pd.to_datetime(df_cleaned['Timestamp'])

# 2. Filter invalid ages
df_cleaned = df_cleaned[(df_cleaned['Age'] >= 18) & (df_cleaned['Age'] <= 100)]

# 3. Clean 'Gender' column
def clean_gender(g):
    g = str(g).strip().lower()
    if g in ['male', 'm', 'man', 'cis male', 'malr', 'maile']:
        return 'Male'
    elif g in ['female', 'f', 'woman', 'cis female', 'femake']:
        return 'Female'
    elif 'trans' in g:
        return 'Trans'
    else:
        return 'Other'

df_cleaned['Gender'] = df_cleaned['Gender'].apply(clean_gender)

# 4. Standardize Yes/No/Maybe responses (strip whitespace and lowercase)
yes_no_cols = [
    'self_employed', 'family_history', 'treatment', 'remote_work', 'tech_company',
    'benefits', 'care_options', 'wellness_program', 'seek_help', 'anonymity',
    'mental_health_consequence', 'phys_health_consequence', 'obs_consequence',
    'coworkers', 'supervisor', 'mental_health_interview', 'phys_health_interview',
    'mental_vs_physical'
]

for col in yes_no_cols:
    df_cleaned[col] = df_cleaned[col].str.strip().str.lower()

# 5. Fill missing values in selected categorical columns with "Don't know" or "Not applicable"
df_cleaned['work_interfere'] = df_cleaned['work_interfere'].fillna("Don't know")
df_cleaned['self_employed'] = df_cleaned['self_employed'].fillna("Don't know")

# 6. Optional: Drop mostly empty 'comments' column
df_cleaned.drop(columns=['comments'], inplace=True)

# Final shape after cleaning
df_cleaned.shape
````

### What all manipulations have you done and insights you found?




**Manipulations:**
1. Converted Timestamp to datetime format for time-based analysis.

2. Removed unrealistic Age values (kept only ages between 18 and 100).

3. Cleaned Gender entries and grouped into Male, Female, Trans, Other.

4. Standardized Yes/No/Maybe columns.

5. Fixed case, spaces, and inconsistent formatting across 18 categorical columns.

6. Handled missing values and filled self_employed and work_interfere with "Don't know".

7. Dropped comments column (mostly empty).

**Initial Insights (Post-Cleaning):**
1. Valid entries reduced from 1259 → 1251 after removing age outliers.

2. Male: Majority of respondents.

3. Female and Other/Trans: Present but less frequent.

4. Missing values handled without dropping rows, preserving overall dataset size.

5. Categorical variables now clean and consistent — ready for visualization.

#### Chart - 11
Remote Work & Work Interference vs. Treatment (Heatmap)

````python
# Create a normalized crosstab
cross_tab = pd.crosstab(index=[df_cleaned['remote_work'], df_cleaned['work_interfere']],
                        columns=df_cleaned['treatment'])

# Normalize the values to show proportions
cross_tab = cross_tab.div(cross_tab.sum(axis=1), axis=0)

# Plot heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(cross_tab, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title('Remote Work & Work Interference vs. Treatment (Proportions)')
plt.xlabel('Sought Treatment')
plt.ylabel('Remote Work / Work Interfere')
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

I pick this plot to visualize how remote work status and work interference relate to treatment behavior.
